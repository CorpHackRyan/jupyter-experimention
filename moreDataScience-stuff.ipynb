```python
# Crosstab
# this means that instead of the previous example, make the gears the columns, and rows are the 'am' value for automatic
pd.crosstab(df['am'], df['gear'])
```

```python
# Sorting entire dataframe based on specific column values
df1 = pd.read_csv('dataset/concat_1.csv')
df2 = pd.read_csv('dataset/concat_2.csv')
df3 = pd.read_csv('dataset/concat_3.csv')
```

```python
dfx = pd.concat([df1,df2]) # by default, it concatenates row wise
# also important to remember you can duplicate the labels in rows, index position cannot (vertical concatenation)
# - multiple things to remember:
#   - by default, vertical concatenation is done.
#   - column names are the same for data frames. 
dfx
```

```python
dfx.loc[[0,1],:]
```

```python
# Concatenate Vertically with different column labels (names)
dfy = pd.concat([df1,df2], axis=1) #axis 1 means forcing it to concatenate column wise, default 0 is row rise
dfy
```

```python
# now if we supply mis matched columns, it will concatenate the columns but add them vertically and horizontally if they don't match
dfz = pd.concat([df1, df3])
dfz
```

```python
# If we supply the axis=1, it will just add the columns and not concatenate horizontally
dfz2 = pd.concat([df1, df3], axis=1)
dfz2
```

```python
# Join multiple DataFrames based on values in Specific columns 
# Mapping Records between two dataframes
# for that we have pd.merge()
# - there is 4 arguments for merge,
# - pd.merge(left, right, left_on, right_on)

# Using accounts.csv and acct_loans.csv
# Find all the account deatils who have taken a loan from our bank
acct_tab = pd.read_csv('dataset/accounts.csv')
loan_tab = pd.read_csv('dataset/acct_loans.csv')
```

```python
acct_tab
```

```python
loan_tab
```

```python
mapped_df = pd.merge(left=acct_tab, right=loan_tab, left_on='accountno', right_on='acno')  # left_on should be a col name from left table, right_on is col name from right table
# This is by default inner join
```

```python
mapped_df
```

```python
# if we want to drop one of the columns, we can do this:
mapped_df.drop('acno', axis=1)
```

```python
mapped_df # you can see the original data did not get changed
```

```python
# now to do a outer join, specificy outer 
mapped_df_outer = pd.merge(left=acct_tab, right=loan_tab, left_on='accountno', right_on='acno', how='outer')  # left_on should be a col name from left table, right_on is col name from right table
```

```python
mapped_df_outer
```
